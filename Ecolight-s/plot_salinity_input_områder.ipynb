## Plotting the input chla in the model to see seasonal changes here, so it can be compared to the seasonal changes in RRS

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data_2023 = pd.read_csv("data/area_values_csv_23.csv", delimiter=';')
data_2025 = pd.read_csv("data/area_values_csv.csv", delimiter=';')
#data_new_2025 = pd.read_csv("data/inputs_new_2025.csv")


# Extract the 'chla_mg_m3' column
chla_values_2023 = data_2023['salinity23'].values
chla_values_2025 = data_2025['salinity25'].values


# Optional: print or inspect the values
print(chla_values_2023)

plt.figure(figsize=(10, 5))
#plt.plot(chla_values_2023, marker='o', linestyle='-', color='green')
plt.plot(chla_values_2025, marker='o', linestyle='-', color='red')

plt.grid(True, alpha=0.3)
plt.title("Chlorophyll-a Concentration Over Samples")
plt.xlabel("Sample Index")
plt.ylabel("Chlorophyll-a (mg/m³)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
